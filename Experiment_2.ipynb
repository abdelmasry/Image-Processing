- choose an image with low contrast whose pixels not occupy the full dynamic range and work with it as gray image.
- deduce the original image range from a to b [min and max of image pixels]
- Plot its histogram
- Apply contrast stretching on image pixels to utilize the full dynamic range [0,255]
- Display the new image
- Plot the modified image histogram

```python
import cv2 as cv
from skimage.io import imread,imshow,imsave
from matplotlib import pyplot as plt
```

```python
img = cv.imread('/Users/abdelrahmanibrahim/Desktop/MASTER/Task_1/test.ppm.png',0)
plt.figure()
plt.axis('off')
imshow(img)
plt.show()
print("MIN INTENSITY: ",img.min())
print("MAX INTENSITY: ",img.max())
```

```python
h = [(img==v).sum() for v in range(256)]

plt.figure()
plt.bar(range(256), h)
plt.show()
```

CONTRAST STRETCHING

```python
from colorama import Fore
import numpy as np

stretched_img = np.zeros((img.shape[0],img.shape[1]),dtype = 'uint8')

# Loop over the image and apply Min-Max formulae
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        stretched_img[i,j] = 255*(img[i,j]-np.min(img))/(np.max(img)-np.min(img))

res = np.hstack((img,stretched_img)) # stacking images side-by-side
cv.imwrite('/Users/abdelmasry/PycharmProjects/ImageProcessing/Task_1/stretched.ppm.png', res)
im = imread('/Users/abdelmasry/PycharmProjects/ImageProcessing/Task_1/stretched.ppm.png')

print(Fore.GREEN+"Stretched image:")
plt.figure()
plt.axis('off')
imshow(im)
plt.show()
print("MIN INTENSITY: ",stretched_img.min())
print("MAX INTENSITY: ",stretched_img.max())

h = [(im==v).sum() for v in range(256)]

print(Fore.GREEN+"Stretched Histogram:")
plt.figure()
plt.bar(range(256), h)
plt.show()
```
