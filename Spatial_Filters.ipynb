```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
from skimage.io import imread,imshow
```

## Experiment-1

### Use image (‘eight.jpg’)
### -Apply Gnoise on the given image
### -Apply different smoothing filters with different window size (Average and median filter)

```python
eight_img = cv.imread('eight.jpg',0)
plt.figure()
plt.axis('off')
imshow(eight_img)
plt.show()
```

```python
def gaussian_noise(img):
   gauss_noise = np.zeros((img.shape[0],img.shape[1]),dtype=np.uint8)
   cv.randn(gauss_noise,128,20)
   gauss_noise=(gauss_noise*0.5).astype(np.uint8)
   gn_img = cv.add(img,gauss_noise)
   return gn_img
```

```python
eight_img_noised = gaussian_noise(eight_img)
cv.imwrite('eight_noise.jpg',eight_img_noised)
plt.figure()
plt.axis('off')
imshow(eight_img_noised)
plt.show()
```

## Average Filter with different kernel sizes

```python
img_blur = cv.blur(eight_img_noised,(7,7))
img_blur_2 = cv.blur(eight_img_noised,(5,5))
img_blur_3 = cv.blur(eight_img_noised,(3,3))
cv.imwrite('smoothed_eight.jpg',img_blur_3)
print("7 by 7: ")
plt.figure()
plt.axis('off')
imshow(img_blur)
plt.show()

print("5 by 5: ")
plt.figure()
plt.axis('off')
imshow(img_blur_2)
plt.show()

print("3 by 3: ")
plt.figure()
plt.axis('off')
imshow(img_blur_3)
plt.show()
```

## Median filter with different kernel sizes

```python
print("5 by 5")
image_median = cv.medianBlur(eight_img_noised,5)
plt.figure()
plt.axis('off')
imshow(image_median)
plt.show()
print("3 by 3")
image_median_3 = cv.medianBlur(eight_img_noised,3)
cv.imwrite('median_eight.jpg',image_median_3)
plt.figure()
plt.axis('off')
imshow(image_median_3)
plt.show()
```

# Experiment-2
### Use image (‘circuit.jpg’)
#### -Apply different sharpening filters. (Sobel and Laplacian Filters)

```python
circuit_img = cv.imread('circuit.jpg',0)
plt.figure()
plt.axis('off')
imshow(circuit_img)
plt.show()
```

## Sobel

```python
sobel_dy = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
sobel_dx = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
dy = cv.filter2D(circuit_img,0,sobel_dy)
sobel = cv.filter2D(dy,-1,sobel_dx)
plt.figure()
plt.axis('off')
imshow(sobel)
plt.show()
```

## Laplacian

```python
laplacian_kernel = np.array([[0,1,0],[1,-4,1],[0,1,0]])
laplacian = cv.filter2D(circuit_img,0,laplacian_kernel)
plt.figure()
plt.axis('off')
imshow(laplacian)
plt.show()
```
